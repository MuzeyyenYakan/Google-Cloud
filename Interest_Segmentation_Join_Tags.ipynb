```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("gs://interest_segmentation/04.Cloud/02.Merchant_Tags.csv", encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
df.rename(columns={'Id': 'merchant_id', 'B_MERCHANT_NAME': 'merchant_name'},inplace=True)
```

```python
df.head()
```

```python
df.sort_values("merchant_id", axis = 0, ascending = True,  inplace = True, na_position ='last') 
```

```python
df.head()
```

```python
df_melt = pd.melt(df, id_vars=['merchant_id', 'merchant_name', 'Sektör'], var_name='i_tag_name', value_name='i_tag_value')
```

```python
df_importedtags = pd.read_csv("gs://interest_segmentation/04.Cloud/00.Imported_Tags.csv", encoding = "ISO-8859-1")
```

```python
df_importedtags.drop(['Unnamed: 2'], axis=1, inplace=True)
```

```python
df_importedtags.rename(columns={'I_Tag_Name': 'i_tag_name'},inplace=True)
```

```python
df_importedtags.head()
```

```python
df_final = pd.merge(df_melt, df_importedtags, on='i_tag_name', how='left')
```
